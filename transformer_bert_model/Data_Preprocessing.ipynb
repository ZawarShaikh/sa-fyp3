!------------------------------------------------- Balancing the dataset of each product ----------------------------------!

here we are using undersampling to balance the data. 
 - we have taken only 11 different products to make balanced dataframe. airCooler_65_df and airCooler_75_df considered to be one because it is related to air cooler only 

```python
# Using groupby to split the dataset by Sentiment type and taking minimum samples from each group
# We will do this process for each of the product dataframe by considering their min Sentiment which are above

# 1- menCargo_df
balanced_menCargo_df = menCargo_df.groupby('Sentiment').apply(lambda x: x.sample(100)).reset_index(drop=True)

# 2- dinnerSet_df
balanced_dinnerSet_df = dinnerSet_df.groupby('Sentiment').apply(lambda x: x.sample(98)).reset_index(drop=True)

# 3- lakmeKajal_df
balanced_lakmeKajal_df = lakmeKajal_df.groupby('Sentiment').apply(lambda x: x.sample(100)).reset_index(drop=True)

# 4- elctricKettle_df
balanced_elctricKettle_df = elctricKettle_df.groupby('Sentiment').apply(lambda x: x.sample(97)).reset_index(drop=True)

# 5- windowCurtain_df
balanced_windowCurtain_df = windowCurtain_df.groupby('Sentiment').apply(lambda x: x.sample(75)).reset_index(drop=True)

# 6- memoryCard_df
balanced_dslrCamera_df = dslrCamera_df.groupby('Sentiment').apply(lambda x: x.sample(70)).reset_index(drop=True)

# 7- blackTshirt_df
balanced_blackTshirt_df = blackTshirt_df.groupby('Sentiment').apply(lambda x: x.sample(99)).reset_index(drop=True)

# 8- dryIron_df
balanced_dryIron_df = dryIron_df.groupby('Sentiment').apply(lambda x: x.sample(100)).reset_index(drop=True)

# 9- cusionPack_df
balanced_cusionPack_df = cusionPack_df.groupby('Sentiment').apply(lambda x: x.sample(96)).reset_index(drop=True)

# 10- googleAssistant_df
balanced_googleAssistant_df = googleAssistant_df.groupby('Sentiment').apply(lambda x: x.sample(77)).reset_index(drop=True)

# 11- airCooler_65_df
balanced_airCooler_65_df = airCooler_65_df.groupby('Sentiment').apply(lambda x: x.sample(34)).reset_index(drop=True)

# 12- airCooler_75_df
balanced_airCooler_75_df = airCooler_75_df.groupby('Sentiment').apply(lambda x: x.sample(40)).reset_index(drop=True)

```

The reason for taking not more than 100 samples for each class is to balance the dataset in such a way that it not affect the model's accuracy, because we also have those products which has min 70 reviews per class. 
So each product has max of 100 reviews per class. 

concatinating our product datasets to make new balanced dataframe

```python
balanced_df = pd.concat([balanced_airCooler_65_df, balanced_airCooler_75_df, balanced_blackTshirt_df, 
                         balanced_cusionPack_df, balanced_dinnerSet_df, balanced_dryIron_df, balanced_dslrCamera_df, 
                        balanced_elctricKettle_df, balanced_googleAssistant_df, balanced_lakmeKajal_df, balanced_menCargo_df, 
                        balanced_windowCurtain_df], axis=0)

balanced_df
```
